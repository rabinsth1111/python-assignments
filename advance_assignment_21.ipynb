#### Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?
**Ans:** : True random numbers are generated from a process that is inherently unpredictable. This could be a physical process, like atmospheric noise, radioactive decay, or electronic noise in a circuit. True random numbers are genuinely unpredictable and provide a high level of randomness.

Pseudo-random numbers are generated by algorithms and are not truly random. They are determined by an initial value called a seed.  

Pseudo-Random Numbers are Considered "Good Enough" because of following reasons:

1. They can be initialized with a seed, making them reproducible, which is useful for debugging and testing.

2. They are computationally less expensive and can generate sequences of numbers quickly.

3. They are suitable for many applications, such as simulations, and statistical sampling.

#### Q5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows ?
**Ans:** In the short term, the behavior of a probability distribution can be somewhat unpredictable due to the influence of random variations. However, as the number of trials increases, the distribution tends to stabilize and approach a more consistent pattern. This phenomenon is described by the law of large numbers and is a fundamental concept in probability and statistics.


The law of large numbers states that as the number of trials or observations increases, the sample mean will converge to the true mean of the population, and the sample proportion will converge to the true probability. In other words, the more data you collect, the more reliable and representative your estimates become.


For example, if you were flipping a fair coin, in the short term, you might observe a few heads or tails in a row, creating some variability. However, as you continue to flip the coin a larger number of times, the proportion of heads and tails should approach 0.5, reflecting the true probability of each outcome for a fair coin.

#### Q6. What kind of object can be shuffled by using random.shuffle ?
**Ans:** The random.shuffle function in Python is designed to shuffle the elements of a mutable sequence.
 Mutable sequences are objects whose elements can be changed after the object is created. Examples of mutable sequences in Python include lists and arrays.

```python
import random

my_list = [1, 2, 3, 4, 5]
random.shuffle(my_list)

print(my_list)
```

random.shuffle cannot be directly applied to immutable sequences such as tuples or strings because they cannot be modified after creation.

```python
random.shuffle((1,4,5))
```

```python
random.shuffle('hello')
```

#### Q7. Describe the math package's general categories of functions ?
**Ans:** The functions in the math module can be broadly categorized into several groups and they are:

**1. Basic Arithmetic Functions:**

math.ceil(x): Returns the smallest integer greater than or equal to x.<br/>
math.floor(x): Returns the largest integer less than or equal to x.<br/>
math.trunc(x): Truncates x to the nearest integer toward 0.<br/>
math.sqrt(x): Returns the square root of x.<br/>
math.pow(x, y): Returns x raised to the power y.<br/>
math.exp(x): Returns the exponential of x.<br/>


**2. Trigonometric Functions:**

math.sin(x), math.cos(x), math.tan(x): Sine, cosine, and tangent of x (where x is in radians).<br/>
math.asin(x), math.acos(x), math.atan(x): Inverse sine, cosine, and tangent, returning values in radians.<br/>


**3. Hyperbolic Functions:**

math.sinh(x), math.cosh(x), math.tanh(x): Hyperbolic sine, cosine, and tangent.


**4. Logarithmic and Exponential Functions:**

math.log(x): Natural logarithm of x.<br/>
math.log10(x): Base-10 logarithm of x.<br/>
math.log2(x): Base-2 logarithm of x.<br/>
math.exp(x): Exponential function (equivalent to e**x).<br/>


**5. Constants:**

math.pi: A mathematical constant representing the ratio of a circle's circumference to its diameter.<br/>
math.e: Euler's number, the base of natural logarithms.<br/>


**6. Angular Conversion:**

math.degrees(x): Converts angle x from radians to degrees.<br/>
math.radians(x): Converts angle x from degrees to radians.<br/>


**7.Other Functions:**

Various mathematical constants like math.inf (positive infinity) and math.nan (representation of "Not a Number").


#### Q8. What is the relationship between exponentiation and logarithms ?
**Ans:** Exponentiation and logarithms are inverse of each other. 

```python
#Example
import math

# Exponentiation
base = 2
exponent = 3
result_exp = math.pow(base, exponent)
print(f"{base}^{exponent} = {result_exp}")

# Logarithms
argument = 8
result_log = math.log(argument, base)
print(f"log_{base}({argument}) = {result_log}")
```
